```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data= "Resources/schools_complete.csv"
student_data= "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
numstu=len(school_data_complete)
numstu
```

```python
school_type=school_data.set_index(["school_name"])["type"]
```

```python
#placing budget after mean, allows for budget to be calculated as a float, due to the fact that 
#you're taking the mean of the entire df, then only looking at the budget column

per_school_budget=school_data_complete.groupby(['school_name']).mean()['budget']
per_school_budget
```

```python
per_school_counts=school_data_complete["school_name"].value_counts()
per_school_counts
```

```python
per_student_budget=(per_school_budget/per_school_counts)
per_student_budget
```

```python
math_mean=school_data_complete.groupby(["school_name"]).mean()["math_score"]
math_mean
```

```python
reading_mean=school_data_complete.groupby(["school_name"]).mean()["reading_score"]
reading_mean
```

```python
passingMath_count = school_data_complete[school_data_complete['math_score'] > 70]
passingMath_count
```

```python
passingRead_count = school_data_complete[school_data_complete['reading_score'] > 70]
passingRead_count
```

```python
per_school_percent_MPass=(passingMath_count.groupby(["school_name"]).count()['student_name']/per_school_counts)*100
per_school_percent_MPass.round()
```

```python
per_school_percent_RPass=(passingRead_count.groupby(["school_name"]).count()['student_name']/per_school_counts)*100
per_school_percent_RPass.round()
```

```python
overall=(per_school_percent_MPass+per_school_percent_RPass)/2
overall.round()
```

```python
School_Sum=pd.DataFrame({ "School Type":school_type, 
                          "Total Students":per_school_counts,
                          "Total School Budget":per_school_budget, 
                          "Per Student Budget":per_student_budget,
                          "Average Math Score":math_mean.round(),
                          "Average Reading Score":reading_mean.round(),
                          "Percentage Passing Math":per_school_percent_MPass.round(),
                          "Percentage Passing Reading":per_school_percent_RPass.round(),
                          "Overall Passing Rate":overall.round().astype(str)+'%'})
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grades_9 = student_data.loc[student_data["grade"] == "9th"]
grades_10 = student_data.loc[student_data["grade"] == "10th"]
grades_11 = student_data.loc[student_data["grade"] == "11th"]
grades_12 = student_data.loc[student_data["grade"] == "12th"]
```

```python
grades_9th=grades_9.groupby(["school_name"]).mean()["math_score"].round()
grades_10th=grades_10.groupby(["school_name"]).mean()["math_score"].round()
grades_11th=grades_11.groupby(["school_name"]).mean()["math_score"].round()
grades_12th=grades_12.groupby(["school_name"]).mean()["math_score"].round()
grades_9th
```

```python
#merge_table = pd.merge(info_pd, items_pd, on="customer_id")
Grade_Summary=pd.merge(grades_9th,grades_10th, on="school_name",how="outer")
Grade_Summary.columns=['9th', '10th']
Grade_Summary
```

```python
Grade_Summary2=pd.merge(grades_11th,grades_12th, on="school_name")
Grade_Summary2.columns=['11th','12th']
Grade_Summary2
```

```python
Grade_Summary_Math=pd.merge(Grade_Summary, Grade_Summary2, on="school_name", how="inner")
Grade_Summary_Math
```
